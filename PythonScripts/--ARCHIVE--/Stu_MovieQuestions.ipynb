```python
# Dependencies
import requests
import json
from pprint import pprint
import pandas as pd

url = "http://www.omdbapi.com/?t="
api_key = "&apikey=trilogy"
```

```python
movies = ["Aliens", "Gladiator", "Sing", "Moana", "50 First Dates"]
responses = []
```

```python
for x in range(len(movies)):
    print(f"Making request number: {x} for ID: {movies[x]}")

    # Get one of the posts
    post_response = requests.get(url + str(movies[x]+api_key))
    
    
    responses.append(post_response.json())
    
    responses
    print(type(responses))

```

```python
pprint(responses)
```

```python
df = pd.DataFrame(responses)
df.head(20)
```
