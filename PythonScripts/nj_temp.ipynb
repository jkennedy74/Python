```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
months = np.arange(1,13,1)
print(months)
```

```python
F = np.array([ 39, 42, 51, 62, 72, 82, 86, 84, 77, 65, 55, 44])
F
```

```python
plt.plot(months, F) 
plt.show()
```

```python
#Convert F to C  Fix the list comprehension.  Something is out of order

C = [ (F - 39) * .56 for F in months]
C
```

```python
plt.plot(months, C) 
plt.show()
```

```python
plt.plot(months, F) 
plt.plot(months, C) 

plt.show()
```
