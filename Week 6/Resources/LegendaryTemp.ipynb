```python
# Dependencies
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Set x axis to numerical value for month
x_axis_data = np.arange(1,13,1)
x_axis_data
```

```python
# Average weather temp
points = [39, 42, 51, 62, 72, 82, 86, 84, 77, 65, 55, 44]
```

```python
# Plot the line
plt.plot(x_axis_data, points, 'b+', lw=1, ls='-')
plt.show()
```

```python
# Convert to Celsius C = (F-32) * 0.56
points_C = [(x-32) * 0.56 for x in points]
points_C
```

```python
# Plot using Celsius
plt.plot(x_axis_data, points_C)
plt.show()
```

```python
# Plot both on the same chart



Line_F, = plt.plot(x_axis_data, points_C, 'rs', lw = 1, ls='-', label='F')
Line_C, = plt.plot(x_axis_data, points, 'b+', lw=1, ls='-', label='C')
plt.legend([Line_F, Line_C], ['Farenheit', 'Celsius'])

plt.show()
```
